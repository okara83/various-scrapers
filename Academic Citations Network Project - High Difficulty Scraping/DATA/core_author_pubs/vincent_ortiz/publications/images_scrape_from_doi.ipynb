```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import re
import json
```

# Open particular researchers 'full_info.xls' file

```python
df = pd.read_excel("full_info.xls")
```

Turn into list

```python
dois = ["https://doi.org/"+str(i) for i in df["DOI"].tolist() if i != "nan"]
print(dois)
```

```python
def returnAbstractImages(url):
    header= {'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_11_5) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/50.0.2661.102 Safari/537.36'}
    r = requests.get(url,headers=header)
    neurl = r.url
    print(neurl)
    url2=neurl
    paltlinks = {}

    links=[]
    if "elsevier.com/retrieve" in neurl:
        url2 ="https://www.sciencedirect.com/science/article/abs"+ re.sub("^.*/pii","/pii",neurl).strip()
        r = requests.get(url2,headers = header)
        soup = BeautifulSoup(r.content, 'lxml')
        abstract_images = soup.find("div", {"class": "abstract graphical"})

        if not abstract_images:
            print(links,"  ... \n... ", "ELV")
            return {url.replace("https://doi.org/","").strip():{"abstract_url":url2,"images":links}}

        abstract_images = abstract_images.findAll('a', {"class":"anchor download-link u-font-sans"})
        for i in abstract_images:
            links.append(i.get('href'))
        #print(links,"  ... \n... ", "ELV")

        paltlinks[url.replace("https://doi.org/","").strip()] ={"abstract_url":url2,"images":links}
        return paltlinks

    else:
        r = requests.get(url,headers = header)
        source = r.content
        soup = BeautifulSoup(r.content, 'lxml') 
        for x in soup.find_all('img', alt= True): 
            if x['alt'] == "Abstract Image": 
                links.append(x.get('src'))
            print(links,"  ... \n... ", "ACS")
    return {url.replace("https://doi.org/","").strip():{"abstract_url":url2,"images":links}}
```

```python
output1=[]
cnt=0
for i in dois:
    print(cnt)
    cnt+=1
    images = returnAbstractImages(i)
    output1.append(images)
```

```python
for i in output1:
   # print(i.keys())
    b = list(i.keys())[0]
    if len(i[b]['images'])>0:
        print(i,i[b]['images'],"\n")
with open("images_doi_all.json","w") as file:
    json.dump(output1,file)
```
