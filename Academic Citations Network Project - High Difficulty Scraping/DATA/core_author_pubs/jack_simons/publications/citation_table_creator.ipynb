```python
for i in c:
    if i[1] and len(str(i[1]))>5:
        embedd1 ='<th><a href="https://doi.org/'+i[1]+'"> '+ i[0]+'</a></th>'
        embedd2 = '<th>'+dummy1+i[1]+ '" data-style="small_rectangle">' + dummy2 +'</th>'
        outgoing_embeddings.append(('<tr>'+embedd1 + embedd2+'</tr>'))
```

```python
with open("test_metric4.html","w") as file:
    file.write('<html><body><div style="overflow-x:auto;"><table style="width:650px">')
    for i in outgoing_embeddings:
        print(i)
        file.write(i)
    file.write("</table></div></body></html>")
```
