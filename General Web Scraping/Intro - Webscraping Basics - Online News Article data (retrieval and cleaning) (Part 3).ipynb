# 'Inspect' original HTML - quickest way to locate desired information is this two-pronged approach

Usually we want content that is not a headline, title, section header, ect... We only want to retrieve coherent proper/full sentence and paragraphs

Upon inspection we see for this particular url on buzzfeed that in information we want is between plain p tags (i.e. no class, or custom identifiers). Also notice in this case we want to remove b tags nested within p tags

# Two ways to retrieve desired content from HTML string
<ol> 
    <li>Regular Expression (RegEx) </li>
    <li>BeautifulSoup</li>
</ol>

Paragraph (p tag) retrieval using regular expressions:

```python
p_tags = re.findall(r"<p>(.*?)</p>",page_content,re.DOTALL) #all content from starting with <p> and ending </p> - NON-GREEDY
print(len(p_tags))
```

```python
p_tags_greedy = re.findall(r"<p>(.*)</p>",page_content,re.DOTALL) #all content from starting with <p> and ending </p> - GREEDY
print(len(p_tags_greedy))
```

```python
print(p_tags)
```

```python
content_dirty = " ".join(p_tags)
```

```python
print(content_dirty)
```

# Tags to keep - i tags, a tags, em tags - used as entity indicators <br>Tags to remove - b tags

```python
content_dirty = " ".join(" ".join(content_dirty.split("\n")).split("\t"))
content_clean = re.sub(r"<b>.*?</b>"," ",content_dirty,re.MULTILINE)
print(content_clean)
```

```python
to_remove = re.findall(r"<b>.*?</b>",content_dirty,re.DOTALL)
print(len(to_remove))
```

```python
for i in to_remove:
    content_clean = content_clean.replace(i, " ")
print(content_clean,"\n\n","<br>" in content_clean)
```

```python
content_clean = " ".join(content_clean.split())
```

```python
print(content_clean)
```

# Bad paragraphs snuck in

We do not want text that is in paragraph tags nested within image captions, figures, block quotes, and iframe, twitter block, etc... So we need to pre-clean before finding all p tags

```python
page = requests.get(url).text
soup = BeautifulSoup(page,'html.parser')
```

```python
content = soup.find('div', class_="content")
content=str(content)
soup = BeautifulSoup(content,'html.parser')
[tagstring.extract() for tagstring in soup('script')] ### extracts all content not between balanced <script> tags
[tagstring.extract() for tagstring in soup('table')] ### extracts all content not between balanced <table> tags
[tagstring.extract() for tagstring in soup('iframe')] ### extracts all content not between balanced <iframe> tags
[tagstring.extract() for tagstring in soup('blockquote')] ###OUR CULPRIT extracts all content not between balanced <blockquote> tags
filtered_content_list = list(soup.find_all('p'))
```

```python
content_list = [str(tag) for tag in filtered_content_list]
```

```python
content_list
```

```python
content = " ".join(content_list)
```

```python
content
```

```python
print(content.encode("utf-8"))#\xa0 is non-breaking space in Latin1(ISO 8859-1), also chr(160),When .encode('utf-8'), it will encode the unicode to utf-8, that means every unicode could be represented by 1 to 4 bytes. For this case, \xa0 is represented by 2 bytes \xc2\xa0.
```
