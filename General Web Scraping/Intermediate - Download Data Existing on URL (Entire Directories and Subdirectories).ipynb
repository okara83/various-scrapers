```python
from urllib.request import urlopen
import re
from bs4 import BeautifulSoup
import requests
import wget

class DownloadDirectorySubdirectory:
    
    def __init__(self,url):
        self.url = url
        self.outputdirectories = []
        self.finalsearches_download=[]
        
    def isDirectory(urlpage):
        if(urlpage.endswith('/')):
            return True
        else:
            return False

    def findLinks(self):
        outputdirectory=[]
        outputdirectory.append(self.url)
        files=[]
        page = requests.get(self.url).content
        bsObj = BeautifulSoup(page, 'html.parser')
        maybe_directories = bsObj.findAll('a', href=True)
        for link in maybe_directories:
            hrefstr= re.search(r'href=\"(.*?)\"',str(link))[1]
            print(hrefstr)
            if hrefstr[0] !="/" and hrefstr[-1]=="/":
                outputdirectory.append(self.url + hrefstr)
            elif len(hrefstr.split("."))==2:
                files.append(hrefstr)
                
            try:
                if(isDirectory(link['href'])):
                    newUrl = self.url + link['href']         
                    links = findLinks(newUrl) #recursion happening here
#                     print(links)
                else:
                    if(link['href'].endswith('maven-metadata.xml')):
                        print("GOTCHA!") #now safe and download
            except:
                continue
        finalsearches_download=[]
        for j in outputdirectory:
            for k in files:
                finalsearches_download.append(j+k)
        self.finalsearches_download = finalsearches_download
        self.outputdirectories = outputdirectory
        return
    
    def download_links(self,save_path=""):
        for link in self.finalsearches_download:
            try:
                wget.download(link)
            except:
                continue
```
